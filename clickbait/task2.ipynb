```python
import jsonlines

d =[]

with jsonlines.open('train.jsonl') as f:

    for line in f.iter():
        d.append(line)
columns  = d[0].keys()
import pandas as pd
df = pd.DataFrame(columns=columns)
for i in d:
    df.loc[len(df.index)] = list(i.values())
```

```python
d =[]

with jsonlines.open('validation.jsonl') as f:

    for line in f.iter():
        d.append(line)
columns  = d[0].keys()
import pandas as pd
valid = pd.DataFrame(columns=columns)
for i in d:
    valid.loc[len(valid.index)] = list(i.values())
```

```python
import logging

from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```python
# Configure the model
model_args = QuestionAnsweringArgs()
model_args.train_batch_size = 25
model_args.evaluate_during_training = True
from simpletransformers.question_answering import QuestionAnsweringModel


model = QuestionAnsweringModel(
    "roberta", "roberta-base"
)
```

```python
df.columns
```

```python
train = []
for i in range(df.shape[0]):
    a = df.loc[i]
    d = {}
    d['context'] = ' '.join(a['targetParagraphs'])
    e = []
    uuid = a['uuid']
    for i in range(len(a['spoilerPositions'])):
        id = uuid +' '+ str(i)
        is_impossible = False
        question = a['targetParagraphs'][a['spoilerPositions'][i][0][0]]
        answers = [{'text':a['spoiler'][i], 'answer_start':d['context'].find(a['spoiler'][i])}]




        e.append( { 'id':id,'is_impossible':is_impossible,'question':question,'answers':answers})
    d['qas'] = e
    train.append(d)
```

```python
test = []
for i in range(valid.shape[0]):
    a = valid.loc[i]
    d = {}
    d['context'] = ' '.join(a['targetParagraphs'])
    e = []
    uuid = a['uuid']
    for i in range(len(a['spoilerPositions'])):
        id = uuid +' '+ str(i)
        is_impossible = False
        question = a['targetParagraphs'][a['spoilerPositions'][i][0][0]]
        answers = [{'text':a['spoiler'][i], 'answer_start':d['context'].find(a['spoiler'][i])}]




        e.append( { 'id':id,'is_impossible':is_impossible,'question':question,'answers':answers})
    d['qas'] = e
    test.append(d)
```

```python
model.train_model(train, eval_data=test)
```

```python
# Evaluate the model
result, texts = model.eval_model(test)
```

```python
result
```

```python
answers, probabilities = model.predict(test)
```

```python
predictions =[]
a = answers[0]['id'].split()[0]
if answers[0]['answer'][0] =='':
  e = [answers[0]['answer'][1]]
else:
  e = [answers[0]['answer'][0]]

for id in answers[1:]:


  if a == id['id'].split()[0]:
    if id['answer'][0] =='':
        e.extend([id['answer'][1]])
    else:
      e.extend([id['answer'][0]])
  else:
    predictions.append(e)
    if id['answer'][0]=='':
       e = [id['answer'][1]]
    else:
      e = [id['answer'][0]]

  a = id['id'].split()[0]
predictions.append(e)




```

```python
uuid = pd.Series(list(valid['uuid']),name='uuid')
predictions =  pd.Series(predictions, name='Spoiler_Prediciton')
```
